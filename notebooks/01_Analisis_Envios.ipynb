Antes de la limpieza, el dataset presentaba varios problemas de calidad de datos. Se identificaron valores nulos en columnas importantes como id_envio, id_ruta, id_vehiculo, peso_kg y volumen_m3, lo que afectaba a la información. Además, existían inconsistencias en los tipos de datos, por ejemplo peso_kg estaba como texto en lugar de numérico y las fechas tenían formatos mixtos. También se detectaron errores en variables categóricas como tipo_carga y estado, donde había valores duplicados por diferencias en mayúsculas y escritura, lo que dificultaba su análisis.

```python
df_envios.boxplot(column="peso_kg")
plt.title("Detección de outliers en peso_kg")
plt.show()
```

```python
Q1 = df_envios["peso_kg"].quantile(0.25)
Q3 = df_envios["peso_kg"].quantile(0.75)
IQR = Q3 - Q1

outliers = df_envios[
    (df_envios["peso_kg"] < Q1 - 1.5 * IQR) |
    (df_envios["peso_kg"] > Q3 + 1.5 * IQR)
]

outliers
```

```python
print("Cantidad de outliers:", outliers.shape[0])
```

```python
# peso a float
df_envios["peso_kg"] = pd.to_numeric(df_envios["peso_kg"], errors="coerce")

# id_envio a int (pero primero manejar nulos)
df_envios["id_envio"] = df_envios["id_envio"].astype("Int64")
```

```python
df_envios["fecha_envio"] = pd.to_datetime(df_envios["fecha_envio"], errors="coerce")
df_envios["fecha_entrega"] = pd.to_datetime(df_envios["fecha_entrega"], errors="coerce")
```

```python
df_envios["tipo_carga"] = df_envios["tipo_carga"].str.lower().str.strip()
df_envios["estado"] = df_envios["estado"].str.lower().str.strip()
```

```python
# eliminar nulos en id
df_envios = df_envios.dropna(subset=["id_envio"])

# rellenar peso correctamente
df_envios["peso_kg"] = df_envios["peso_kg"].fillna(df_envios["peso_kg"].mean())
```

```python
Q1 = df_envios["peso_kg"].quantile(0.25)
Q3 = df_envios["peso_kg"].quantile(0.75)
IQR = Q3 - Q1

filtro = (df_envios["peso_kg"] >= Q1 - 1.5 * IQR) & (df_envios["peso_kg"] <= Q3 + 1.5 * IQR)

df_envios = df_envios[filtro]
```

Luego del proceso de limpieza, se eliminaron los registros sin id_envio, se imputaron valores faltantes en peso_kg utilizando la media y se corrigieron los tipos de datos, transformando las fechas a formato datetime y las variables numéricas a su tipo correspondiente. Además, se estandarizaron las variables categóricas convirtiéndolas a minúsculas y unificando sus valores. Con esto, el dataset quedó consistente, sin errores estructurales importantes y preparado para las siguientes etapas de transformación y análisis.
