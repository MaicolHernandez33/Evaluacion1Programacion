## 6. Revisión de valores nulos

Se analiza la cantidad y porcentaje de valores faltantes por columna. Esto permite identificar qué variables podrían requerir tratamiento adicional antes del modelado.

```python
nulos = pd.DataFrame({
    "nulos": df.isnull().sum(),
    "porcentaje": (df.isnull().mean() * 100).round(2)
})

nulos = nulos.sort_values("nulos", ascending=False)
nulos
```

```python
plt.figure(figsize=(12, 6))
nulos_con_valores = nulos[nulos["nulos"] > 0]

sns.barplot(
    data=nulos_con_valores.reset_index(),
    x="porcentaje",
    y="index"
)

plt.title("Porcentaje de valores nulos por columna")
plt.xlabel("Porcentaje de nulos (%)")
plt.ylabel("Columnas")
plt.tight_layout()
plt.show()
```

### Interpretación

Las variables con valores faltantes restantes corresponden principalmente a atributos descriptivos provenientes de los joins entre envíos, rutas y vehículos. Las variables críticas para modelado, como `entrega_tardia`, `dias_entrega`, `peso_kg`, `id_envio`, `id_ruta` e `id_vehiculo`, deben mantenerse completas para garantizar continuidad en la etapa de Machine Learning.

## 8. Variable objetivo para clasificación: entrega tardía

La variable `entrega_tardia` permite identificar si un envío fue entregado tarde o no. Esta será una variable clave para la EVA 2.

```python
df["entrega_tardia"].value_counts()
```

```python
plt.figure(figsize=(7, 5))

sns.countplot(
    data=df,
    x="entrega_tardia"
)

plt.title("Distribución de entregas tardías")
plt.xlabel("Entrega tardía (0 = No, 1 = Sí)")
plt.ylabel("Cantidad de envíos")
plt.tight_layout()
plt.show()
```

```python
proporcion_tardia = df["entrega_tardia"].value_counts(normalize=True).mul(100).round(2)
proporcion_tardia
```

### Interpretación

Este gráfico permite observar si el problema de clasificación está balanceado o desbalanceado. Si una clase domina demasiado, en EVA 2 será importante considerar métricas como `recall`, `precision` y `f1-score`, no solo accuracy.

## 10. Análisis general de incidencias

Se revisa cuántos envíos presentan incidencias y cómo se distribuye esta variable.

```python
df["tiene_incidencia"].value_counts()
```

```python
plt.figure(figsize=(7, 5))

sns.countplot(
    data=df,
    x="tiene_incidencia"
)

plt.title("Distribución de envíos con incidencias")
plt.xlabel("Tiene incidencia (0 = No, 1 = Sí)")
plt.ylabel("Cantidad de envíos")
plt.tight_layout()
plt.show()
```

## 11. Relación inicial entre incidencias y entregas tardías

```python
tabla_incidencia_retraso = pd.crosstab(
    df["tiene_incidencia"],
    df["entrega_tardia"],
    normalize="index"
).round(3)

tabla_incidencia_retraso
```

```python
tabla_plot = tabla_incidencia_retraso.reset_index().melt(
    id_vars="tiene_incidencia",
    var_name="entrega_tardia",
    value_name="proporcion"
)

plt.figure(figsize=(8, 5))
sns.barplot(
    data=tabla_plot,
    x="tiene_incidencia",
    y="proporcion",
    hue="entrega_tardia"
)

plt.title("Relación entre incidencias y entregas tardías")
plt.xlabel("Tiene incidencia")
plt.ylabel("Proporción")
plt.legend(title="Entrega tardía")
plt.tight_layout()
plt.show()
```

### Interpretación

Este análisis permite observar si la presencia de incidencias está asociada a una mayor proporción de entregas tardías. Si la diferencia es relevante, `tiene_incidencia` puede ser una variable importante para los modelos de clasificación.

## 12. Distribución de variables numéricas principales

```python
columnas_numericas_interes = [
    "peso_kg",
    "volumen_m3",
    "distancia_km",
    "tiempo_estimado_hrs",
    "capacidad_kg",
    "capacidad_m3",
    "cantidad_incidencias",
    "costo_total_incidencias",
    "velocidad_promedio_km_h",
    "uso_capacidad_kg",
    "uso_capacidad_m3",
    "dias_entrega"
]

columnas_numericas_interes = [col for col in columnas_numericas_interes if col in df.columns]
columnas_numericas_interes
```

```python
for col in columnas_numericas_interes:
    plt.figure(figsize=(8, 4))
    sns.histplot(data=df, x=col, bins=30, kde=True)
    plt.title(f"Distribución de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()
```

## 13. Boxplots de variables clave según entrega tardía

Se comparan algunas variables numéricas contra `entrega_tardia` para observar diferencias entre entregas normales y tardías.

```python
variables_boxplot = [
    "distancia_km",
    "peso_kg",
    "uso_capacidad_kg",
    "cantidad_incidencias",
    "costo_total_incidencias",
    "dias_entrega"
]

variables_boxplot = [col for col in variables_boxplot if col in df.columns]

for col in variables_boxplot:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, x="entrega_tardia", y=col)
    plt.title(f"{col} según entrega tardía")
    plt.xlabel("Entrega tardía")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

## 14. Matriz de correlación

Se revisan correlaciones entre variables numéricas para detectar relaciones relevantes y posibles variables predictoras.

```python
columnas_corr = df.select_dtypes(include=[np.number]).columns

plt.figure(figsize=(14, 10))
sns.heatmap(
    df[columnas_corr].corr(),
    cmap="coolwarm",
    center=0,
    annot=False
)

plt.title("Matriz de correlación de variables numéricas")
plt.tight_layout()
plt.show()
```
