```python
import matplotlib.pyplot as plt

df_rutas.boxplot(column="distancia_km")
plt.show()
```

```python
Q1 = df_rutas["distancia_km"].quantile(0.25)
Q3 = df_rutas["distancia_km"].quantile(0.75)
IQR = Q3 - Q1

outliers = df_rutas[
    (df_rutas["distancia_km"] < Q1 - 1.5 * IQR) |
    (df_rutas["distancia_km"] > Q3 + 1.5 * IQR)
]

len(outliers)
```

```python
# 1. Eliminar filas sin id (clave primaria)
df_rutas = df_rutas.dropna(subset=["id_ruta"])

# 2. Convertir id a int
df_rutas["id_ruta"] = df_rutas["id_ruta"].astype(int)

# 3. Rellenar nulos numéricos con media
df_rutas["distancia_km"] = df_rutas["distancia_km"].fillna(df_rutas["distancia_km"].mean())
df_rutas["tiempo_estimado_hrs"] = df_rutas["tiempo_estimado_hrs"].fillna(df_rutas["tiempo_estimado_hrs"].mean())
df_rutas["peaje_total"] = df_rutas["peaje_total"].fillna(df_rutas["peaje_total"].mean())

# 4. Rellenar nulos categóricos
df_rutas["origen"] = df_rutas["origen"].fillna("desconocido")
df_rutas["destino"] = df_rutas["destino"].fillna("desconocido")
df_rutas["tipo_via"] = df_rutas["tipo_via"].fillna("desconocido")

# 5. Limpiar strings
df_rutas["origen"] = df_rutas["origen"].str.lower().str.strip()
df_rutas["destino"] = df_rutas["destino"].str.lower().str.strip()
df_rutas["tipo_via"] = df_rutas["tipo_via"].str.lower().str.strip()

# 6. Eliminar duplicados
df_rutas = df_rutas.drop_duplicates()
```

Se realizó un análisis de outliers sobre la variable distancia_km mediante un boxplot, donde se identificó la presencia de un valor atípico significativamente mayor al resto de los datos. Sin embargo, este valor podría representar una ruta real de larga distancia dentro del sistema logístico, por lo que no se consideró adecuado eliminarlo sin mayor contexto. En este caso, se decidió mantener el dato y solo documentar su existencia, ya que no afecta de forma crítica el análisis general del dataset.
