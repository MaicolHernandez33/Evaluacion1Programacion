El dataset de vehículos contiene 61 registros y 8 columnas, mezclando variables numéricas (float64) y categóricas (object). Se observa la presencia de valores nulos en varias columnas importantes: id_vehiculo, placa, tipo, capacidad_kg, capacidad_m3, año_fabricacion, estado_vehiculo y km_recorridos. Esto indica que algunos registros están incompletos y requieren limpieza. Además, id_vehiculo está como float64 cuando debería ser un identificador entero, y las columnas categóricas como tipo y estado_vehiculo pueden tener inconsistencias en mayúsculas, espacios u otros formatos. Antes de cualquier análisis es necesario corregir los tipos de datos, completar los valores faltantes y estandarizar los strings.

```python
import matplotlib.pyplot as plt

# Boxplots
df_vehiculos.boxplot(column=["capacidad_kg", "capacidad_m3", "km_recorridos"], rot=45)
plt.show()

# O IQR para identificar outliers
for col in ["capacidad_kg", "capacidad_m3", "km_recorridos"]:
    Q1 = df_vehiculos[col].quantile(0.25)
    Q3 = df_vehiculos[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df_vehiculos[(df_vehiculos[col] < Q1 - 1.5*IQR) | (df_vehiculos[col] > Q3 + 1.5*IQR)]
    print(f"{col} tiene {len(outliers)} outliers")
```

```python
# 1. Eliminar filas sin id_vehiculo
df_vehiculos = df_vehiculos.dropna(subset=["id_vehiculo"])

# 2. Convertir id a int
df_vehiculos["id_vehiculo"] = df_vehiculos["id_vehiculo"].astype(int)

# 3. Rellenar nulos numéricos con la media
df_vehiculos["capacidad_kg"] = df_vehiculos["capacidad_kg"].fillna(df_vehiculos["capacidad_kg"].mean())
df_vehiculos["capacidad_m3"] = df_vehiculos["capacidad_m3"].fillna(df_vehiculos["capacidad_m3"].mean())
df_vehiculos["año_fabricacion"] = df_vehiculos["año_fabricacion"].fillna(df_vehiculos["año_fabricacion"].mean())
df_vehiculos["km_recorridos"] = df_vehiculos["km_recorridos"].fillna(df_vehiculos["km_recorridos"].mean())

# 4. Rellenar categóricos
df_vehiculos["placa"] = df_vehiculos["placa"].fillna("desconocida")
df_vehiculos["tipo"] = df_vehiculos["tipo"].fillna("desconocido")
df_vehiculos["estado_vehiculo"] = df_vehiculos["estado_vehiculo"].fillna("desconocido")

# 5. Limpiar strings
for col in ["placa", "tipo", "estado_vehiculo"]:
    df_vehiculos[col] = df_vehiculos[col].str.lower().str.strip()

# 6. Eliminar duplicados
df_vehiculos = df_vehiculos.drop_duplicates()
```

Después de la limpieza, los valores nulos han sido tratados y las columnas categóricas han sido estandarizadas. id_vehiculo ahora es un entero, los duplicados han sido eliminados y las variables numéricas tienen sus valores faltantes rellenados con la media. Se realizó un análisis de outliers y, aunque se identificaron algunos valores extremos, se decidió mantenerlos por ser plausibles dentro del contexto logístico.
